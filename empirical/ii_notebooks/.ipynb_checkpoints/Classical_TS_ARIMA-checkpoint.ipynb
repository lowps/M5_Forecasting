# Checking for Stationarity

1) Plotting rolling mean and variance is a good statistic to see if our series is stationary or non-stationary.

* If rolling statistic show a clear trend (upward or downward)-mean and show varying variance (increasing or decresing amplitude)-std then, a series is possibly non-stationary




```python
fig, ax = plt.subplots(1, figsize=(12,5), tight_layout=False)
ax.plot(ex.index, ex.sales, label="raw data")
#Window 12 chosen due to 12-period seasonality
ax.plot(ex.sales.rolling(window=12).mean(), label="rolling mean", color="k")
ax.plot(ex.sales.rolling(window=12).std(), label="rolling std (x10)", color="g")
ax.legend(bbox_to_anchor=(1.01,1.01))
plt.show()
```

* Visual Conclusion: Non-Stationary, both mean and variance are varying
* ACF Plot: Useful for identifying non-stationary TS. 
    * STATIONARY(ACF): TS, the ACF will drop to zero relatively quickly.
    * NON-STATIONARY(ACF): Data decreases slowly. Also, the value of r1 is often large and positive.
* Unit Root Test: These are statistical hypothesis tests of stationarity that are designed for determining whether differencing is required
* Differencing: Technique to turn a non-stationary TS into stationary; compute the differences between consecutive observations
* Differencing VS Transformations:
    * Transformations (ex: logarithms): Help to stabalise the variance of a time series
    * Differencing: Help stabalise the mean of TS by removing change in the level of a TS, and therefore eliminating (or reducing) trend & seasonality
* Statistical Test to be applied: Augmented Dickey-fuller Test
    * Generates test-statistic with different levels or percentage of confidence.
    * NULL HYPOTHESIS: a unit root exists in the time series, meaning the data is NON-STATIONARY; essentially, it states that the time series has a unit root and is not stationary. 
    * ALTERNATIVE HYPOTHESIS: time series is stationary or trend-stationary, meaning there is no unit root. 

```python
from statsmodels.tsa.stattools import adfuller
```

```python
def adfuller_test(data):
    print("Is the data stationary?")
    dftest = adfuller(data, autolag='AIC')
    print("Test statistic = {:.3f}".format(dftest[0]))
    print("P-value = {:.3f}".format(dftest[1]))
    print("Critical values:")
    for k, v in dftest[4].items():
        print("\t{}: {} - The data is {} stationary with {}% confidence".format(k,v,"not" if v<dftest[0] else "",100-int(k[:-1])))
```

```python
adfuller_test(ex.sales)
print("ADF Statistic > CV 1%")
print("ADF Statistic > CV 5%")
print("ADF Statistic > CV 10%")
print("ADF Statistic is > all CV thus FAIL TO REJECT NULL HYP.")
print("ADF statistic failed to CROSS OVER the CV threshold")
print("P-value 0.237 also confirms: FAIL TO REJECT NULL HYP.")
```

```python
#P-value = 0.237
    #Fail to reject hypothesis, thus suggesting that the null hyp.
    #TS data is NON-STATIONARY, no significant evidence to deny 
    #this statement.
#Test Statistic:
    #A numerical value (e.g., t-score, z-score, chi-square) calculated from your sample data.
    #This is the observed result of your test.
#p-value:
    #The probability of observing your test statistic (or more extreme values) 
    #if the null hypothesis is true.
    #Not the observed result itself, but A MEASURE OF HOW EXTREME THE OBSERVE RESULT IS.
    #Example: A p-value of 0.01 means theres a 1% chance of seeing 
        #your test statistic (or more extreme) under H.
#Significance Level ():
    #The threshold for rejecting H (e.g.,  = 0.05).
#Why the Test Statistic Is the "Observed Result"
    #The test statistic quantifies the observed effect in your data.
    #Its directly tied to your sample and the hypothesis being tested.
    #Example:
        #If you run a t-test and get t = 3.2, 3.2 is your observed result.
        #This value is compared to a critical value or used to calculate the p-value.

#Example Workflow
#1)Observed Data: Sample means of two groups (e.g., Group A = 50, Group B = 55).

#2)Test Statistic (observed result):
    #Calculate a t-score (e.g., t = 2.5) based on the difference in means, sample size, and variability.

#3)Compare to Null Distribution:
    #Determine how likely this t-score is under H (e.g., p = 0.03).

#4)Conclusion:
    #If p <  (e.g., 0.03 < 0.05), reject H.

    
```

```python
#The test statistic in the Augmented Dickey-Fuller (ADF) test 
#is a specialized t-statistic derived from the following regression model:

#Null Hypothesis 
    #The time series has a unit root (i.e., it is non-stationary).
#Alternative Hypothesis 
    #The time series is stationary (no unit root).
#Test Statistic
    #The t-statistic for the coefficient  (GAMMA) in the regression:
    #ADF STATISTIC = Gamma-hat/standard Error(Gamma-hat)
#gamma-hat:
    #Estimated coefficient of yt-1 (lagged level term)
    #This statistic follows the Dickey-Fuller distribution 
    #(not the standard Students t-distribution).
#Interpretation:
    #More negative values of the ADF statistic  Stronger evidence 
        #against the null hypothesis (i.e., stationarity is supported).
    #Compare the ADF statistic to critical values 
    #(e.g., -3.5, -2.9, -2.6 at 1%, 5%, 10% significance levels).
    #if ADF Statistic < Critical Value, reject Ho (conclude stationarity).
#EXAMPLE:
    #ADF Statistic = -4.2
    #Critical Value (1%) = -3.5
    #Since 4.2 < 3.5, reject Ho and conclude the series is stationary
#WHY "Augmented"?
    #The ADF test extends the original Dickey-Fuller test by including lagged differences 
    #to control for autocorrelation in residuals, improving robustness.





```

# Phase 1: Handle Trend Component

* Handling Trend Component of the series 
* Apply Time-varying rolling standardization: Use rolling-window-mean for calculation then divide by rolling-window-std (standard deviation) to de-trend the series
* Technique is used as an attempt to effectively remove TREND component that changes over time, thus de-trending the series.
* 12-day Rolling window used for rolling-mean due to ... INVESTIGATE LATER
* De-trending can help stabilise the MEAN of a time series by removing changes in the LEVEL of a time series, and therefore eliminating (or reducing) trend and seasonality

# Phase 2: Handle Seasonality Component

* De-seasonalizing the series:
* Apply Seasonal Differencing
* Data displays 12-month seasonality, thus subtracing the series with a 12-period-lag difference will handle seasonality by eliminating (or reducing) seasonality component.
* We apply a 12 period lag to remove seasonality from the detrended series

```python
#lagging the series by 12 periods (ex: days, months, etc. look at your data's time stamp)
#in our case, its shifting it by 12 day periods.
#shift(12): retrieves the value from t-12
ex["deseasonal_sales"] = ex["detrended_sales"] - ex["detrended_sales"].shift(12)
```

# ARIMA MODEL

* ARIMA MODEL: AutoRegressive Integrated Moving Average model
* The model consists of 3x parameters
* p: AR parameter
* d: Differencing parameter
* q: MA parameter



* Technique: Grid Search
* Grid Search technique to find the best parameters for ARIMA MODEL

```python
import statsmodels.api as sm
```

```python
p=d=q=range(0,2)
pdq = list(itertools.product(p,d,q))
seasonal_pdq = [(x[0], x[1], x[2], 12) for x in pdq]
aics = []
params = []

for param in pdq:
    for param_seasonal in seasonal_pdq:
        data = ex['sales'].astype('int64').values
        model = sm.tsa.statespace.SARIMAX(data, order=param, seasonal_order=param_seasonal)
        result = model.fit()
        aics.append(result.aic)
        params.append((param, param_seasonal))
        print(f'ARIMA{param}x{param_seasonal}12 - AIC Score:{result.aic}')

index = aics.index(min(aics))
print("Minimum values: " + str(params[index]) + str(aics[index]))
```

```python
def AIC_PDQS(df):
    p = d = q = range(0,2)
    pdq = list(itertools.product(p, d, q))
    pdqs = [(x[0], x[1], x[2], 12) for x in list(itertools.product(p, d, q))]
    
    ans = []
    for comb in pdq:
        for combs in pdqs:
            try:
                mod = sm.tsa.statespace.SARIMAX(df,
                                               order=comb,
                                               seasonal_order=combs,
                                               enforce_stationarity=False,
                                               enforce_invertibility=False)
                output = mod.fit()
                ans.append([comb, combs, output.aic])
                print(f"ARIMA {comb} x {combs}12 : AIC Calculated = {output.aic}")
            except:
                continue
    ans_df = pd.DataFrame(ans, columns=["pdq", "pdqs", "aic"])
    print("MINIMUM VALUE:")
    print(ans_df.loc[ans_df["aic"].idxmin()])
    return ans_df
```

```python
data = ex['sales'].astype('int64').values
ans_df = AIC_PDQS(data)
```

```python
print(ans_df.loc[ans_df["aic"].idxmin()])
#This is actually a better AIC score.
#A lower AIC valus is considered better when comparing different statistical models;
    #it indicates a better balance between model fit & complexity, meaning the model
    #With the LOWEST AIC is generally preferred
```
