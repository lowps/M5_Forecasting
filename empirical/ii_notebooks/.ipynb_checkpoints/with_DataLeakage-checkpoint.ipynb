# Feature Engineering

```python
dfcp["sales_1_day_lag"] = dfcp.groupby("item")["sales"].shift(1)
```

```python
dfcp["sales_7_day_lag"] = dfcp.groupby("item")["sales"].shift(7)
```

```python
dfcp["rolling_14_day_sum"] = dfcp.groupby(["item"])["sales"].transform(lambda x: x.rolling(window=14, min_periods=1).sum())
dfcp["rolling_28_day_sum"] = dfcp.groupby(["item"])["sales"].transform(lambda x: x.rolling(window=28, min_periods=1).sum())
```

```python
#dfcp["rolling_2_day_sum"] = dfcp.groupby(["item"])["sales"].transform(lambda x: x.rolling(window=2, min_periods=2).sum())
```

```python
dfcp
```

```python
dfcp['expanding_mean'] = dfcp.groupby(['item'])['sales'].transform(lambda x: x.expanding().mean().round(2))
```

```python
#dfcp.loc[dfcp["item"] == 9][["date","store","item","sales","rolling_2_day_sum", "expanding_mean"]].head()
```

```python
#dfcp.loc[dfcp["item"]==1].head(50)
```

```python
dfcp["day"] = df["date"].dt.day
dfcp["month"] = df["date"].dt.month
dfcp["year"] = df["date"].dt.year
```

```python
#Monday=0, Sunday=6
dfcp["day_of_week"] = dfcp["date"].dt.weekday
dfcp["weekend"] = dfcp[["day_of_week"]] >=5
dfcp["quarter"] = dfcp["date"].dt.quarter
```

```python
dfcp.dropna(subset=["sales_7_day_lag"],inplace=True)
```

```python
#convert weekend to boolean for models
dfcp['weekend'] = dfcp['weekend'].astype(int)
```

```python
dfcp.loc[dfcp["item"]==9].head()
```

# Model Training

# Global Models: Linear Regression, Random Forest, Prophet

# Linear Regression

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

from sklearn import metrics
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_percentage_error
```

```python
#Split Data for Training
X = dfcp[['item', 'sales_1_day_lag', 'sales_7_day_lag',
       'rolling_14_day_sum', 'rolling_28_day_sum', 'expanding_mean', 'day',
       'month', 'year', 'day_of_week', 'weekend', 'quarter']]

y = dfcp[["sales"]]
```

```python
X_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)
```

```python
X_train.isna().sum()
```

```python
lr_model = LinearRegression()
lr_model.fit(X_train, y_train)
```

```python
y_pred = lr_model.predict(x_test)
```

```python
plt.figure(figsize=(10,6))
plt.scatter(y_test, y_pred)
plt.title("Predicted vs. Ground Truth Sales", loc="left", fontweight="bold")
plt.xlabel("Ground Truth", labelpad = 30)
plt.ylabel("Predicted", labelpad = 30)
    
plt.show()

print("RMSE: ",np.sqrt(metrics.mean_squared_error(y_test, y_pred)).round(2))

print ("MAPE: ",round((mean_absolute_percentage_error(y_test, y_pred)*100),2))
```

```python
custom_order = list(range(1, 51))
sorted_items = sorted(dfcp['item'].unique(), key=lambda x: custom_order.index(x))
for item in sorted_items:
    item_data = dfcp[dfcp['item'] == item].set_index('date')['sales']
    
    # Plot ACF/PACF for each SKU
    fig, axes = plt.subplots(2, 1, figsize=(12, 8))
    plot_acf(item_data, lags=30, ax=axes[0], title=f'ACF for {item}')
    plot_pacf(item_data, lags=30, ax=axes[1], title=f'PACF for {item}')
    
     # Adjust titles: left-align and bold
    for ax in axes:
        title = ax.get_title()  # Get the current title text
        ax.set_title(title,fontweight='bold', fontsize=16)  # Reapply with styling
        
    
    
    plt.tight_layout()
    plt.show()
```

```python
print("Seashell/Scalloped pattern as lag increases.\nSharp spikes at lag 7 & lag 14 (multiples of 7).\nThe decrease in the ACF as the lag increase is due to TREND,\nwhile the Scalloped pattern is due to the SEASONALITY.")
```

```python
#	Note, ACF plot in figure 2.16 is Scalloped Shape and slowly decreasing this signifies: 
    #1) Trend Presence (slow decrease)
    #& 2) Seasonality Presence (scalloped shape)
```

```python
#o	ACF PLOT & its usefulness for identifying non-stationary TS.
#	Stationary TS: The ACF will drop to ZERO (0) relatively quickly, 

#•	ACF Application:
#o	1) ACF on RAW Data: Model selection, identify parameters to use EX: Arima p, q, d.
#o	2) ACF on Residuals: Diagnostic purposes, checks if it captured all of the components in respects to trend, seasonality, etc.
```

```python
#Slow decayiung PACF (vs) Sharp Cut off PACF
#Slow Decaying: Suggests long-term dependency in the data, seasonal pattern in data
#Sharp Cutoff: Abrupt cut off after specific lag suggests direct dcorrelation at 
    #specified lag, which could be tied to a seasonal pattern, ex: lag 12 indicates
    #strong seasonal effect at that interval
#key differences
    #slowdecay- suggests a broad, persistent seasonal influence affecting multiple lags
    #sharp cutoff- suggest specific, strong seasonal effect at a particular/specified lag
    
```

```python
#•	ES & ARIMA models are the two MOST WIDELY used approaches in TS forecasting
#o	ES & ARIMA are complementary methods for TS forecasting.
#o	ARIMA: Focuses on autocorrelations in the data.
    #o	Better for long-term forecasts, and can handle complex patterns & stationary data
#o	ES: Prioritizes trend & seasonality
    #o	Preferred for shorter-term predictions and data w/ clear trends & seasonality
```

```python
#Stationary TS: The TS properties (patterns) do NOT DEPEND on time. 
    #The reason they are NOT dependent on Time is because the constant mean, 
    #constant autocorrelation, constant variance DO NOT change over time thus 
    #not dependent. These properities will look the same at any point in time.
#Non-Stationary TS: The TS properties (patterns) trends & seasonality indicate 
    #NOT stationary or non-stationary because these properties are NOT constant and 
    #they change over time. 
    #o	Non-stationary TS data is DEPENDENT on time.
    #o	Non-stationary properities such as mean, autocorrelation, variance change over time.
    #o	Trend & Seasonality are functions of time, meaning they emerge because of the passage of time.

#	Focus on constant statistical properties and you should always be able to 
    #correctly conclude Stationary (vs) Non-stationary
```

```python
#o	Differencing: A technique to turn non-stationary TS into stationary.
#•	Second-order Differencing
#•	Seasonal Differencing
```

```python
#	Note, You typically apply differencing on the original data to remove trends 
#and stabilize the mean, then you would apply logarithm to the differenced data 
#to stabilize the variance. This sequential approach allows you to handle both issues, 
#thus not applied independently on 2x sets of original TS data.
```

```python
#o	Formal test that objectively determines if differencing is required on the TS, 
    #unit root test.
#o	Kwiatkowski-Phillips_Schmidt-Shin (KPSS) test is utilized in Rob H. TS book.
```
